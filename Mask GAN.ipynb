```python
import json
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from labelme import utils

# Paths to the folders
annotation_folder = '/content/drive/MyDrive/Hood_Masked/'
mask_output_folder = '/content/drive/MyDrive/Image_Masked/'
image_folder = '/content/drive/MyDrive/data/'

# Ensure the output folder exists
if not os.path.exists(mask_output_folder):
    os.makedirs(mask_output_folder)

# Function to create a mask from the polygons
def create_mask(image_shape, polygons):
    mask = np.zeros(image_shape[:2], dtype=np.uint8)
    for polygon in polygons:
        points = np.array(polygon['points'], dtype=np.int32)
        cv2.fillPoly(mask, [points], 255)
    return mask

# Process each JSON annotation file
count = 0
for subfolder in os.listdir(annotation_folder):
    subfolder_path = os.path.join(annotation_folder, subfolder)
    output_subfolder_path = os.path.join(mask_output_folder, subfolder)
    if not os.path.exists(output_subfolder_path):
        os.makedirs(output_subfolder_path)

    for filename in os.listdir(subfolder_path):
        if filename.endswith('.json'):
            filepath = os.path.join(subfolder_path, filename)
            with open(filepath, 'r') as file:
                data = json.load(file)

            # Extract image info and shapes
            image_path = os.path.join(image_folder, subfolder, data['imagePath'])
            image = utils.img_b64_to_arr(data['imageData'])
            image_shape = image.shape
            shapes = data['shapes']

            # Create a mask for each segment
            for shape in shapes:
                label = shape['label']
                mask = create_mask(image_shape, [shape])
                mask_filename = f"{os.path.splitext(filename)[0]}_{label}_mask.jpg"
                mask_filepath = os.path.join(output_subfolder_path, mask_filename)
                success = cv2.imwrite(mask_filepath, mask)
                if not success:
                    print("Failed to save the image.")
                else:
                    print("Image saved successfully.")
                print(f"Mask saved to: {mask_filepath}")

            count += 1

print(f"Total Processed files: {count}")
```
