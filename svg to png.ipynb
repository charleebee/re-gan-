```python
import os
import numpy as np
import cairosvg
from PIL import Image
import io
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Reshape, Flatten, Dropout, BatchNormalization, Activation, ZeroPadding2D, LeakyReLU, UpSampling2D, Conv2D
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
```

```python
def svg_to_image(svg_path, img_size=(128, 128)):
    # Convert SVG to PNG
    png_data = cairosvg.svg2png(url=svg_path)

    # Open PNG image from bytes
    img = Image.open(io.BytesIO(png_data)).convert('RGB')

    # Resize image
    img = img.resize(img_size)

    return np.array(img)

def load_svg_images_from_directory(directory, img_size=(128, 128)):
    images = []
    labels = []
    label_map = {}
    label_counter = 0

    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".svg"):
                file_path = os.path.join(subdir, file)

                # Convert SVG to image
                img = svg_to_image(file_path, img_size)
                images.append(img)

                # Use the directory name as the label
                label_name = os.path.basename(subdir)
                if label_name not in label_map:
                    label_map[label_name] = label_counter
                    label_counter += 1
                labels.append(label_map[label_name])

    images = np.array(images) / 255.0  # Normalize images to [0, 1]
    labels = tf.keras.utils.to_categorical(labels, len(label_map))  # One-hot encode labels

    return images, labels

# Directory containing your SVG files
directory = "/content/drive/MyDrive/SVG/ImposedImages"

# Load SVG images
X_train, y_train = load_svg_images_from_directory(directory, img_size=(128, 128))
```

```python
def build_generator():
    noise_shape = (100,)
    label_shape = (len(y_train[0]),)  # Number of classes from one-hot encoded labels

    noise = Input(shape=noise_shape)
    label = Input(shape=label_shape)

    model_input = tf.keras.layers.concatenate([noise, label])

    x = Dense(256)(model_input)
    x = LeakyReLU(alpha=0.2)(x)
    x = BatchNormalization(momentum=0.8)(x)
    x = Dense(512)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = BatchNormalization(momentum=0.8)(x)
    x = Dense(1024)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = BatchNormalization(momentum=0.8)(x)
    x = Dense(np.prod((128, 128, 3)), activation='tanh')(x)
    img = Reshape((128, 128, 3))(x)

    model = Model([noise, label], img)
    model.summary()

    return model

generator = build_generator()
```

```python
def build_discriminator():
    img_shape = (128, 128, 3)
    label_shape = (len(y_train[0]),)

    img = Input(shape=img_shape)
    label = Input(shape=label_shape)

    label_embedding = Dense(np.prod(img_shape))(label)
    label_embedding = Reshape(img_shape)(label_embedding)

    model_input = tf.keras.layers.concatenate([img, label_embedding])

    x = Conv2D(32, kernel_size=3, strides=2, padding='same')(model_input)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.25)(x)
    x = Conv2D(64, kernel_size=3, strides=2, padding='same')(x)
    x = ZeroPadding2D(padding=((0,1),(0,1)))(x)
    x = BatchNormalization(momentum=0.8)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.25)(x)
    x = Conv2D(128, kernel_size=3, strides=2, padding='same')(x)
    x = BatchNormalization(momentum=0.8)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.25)(x)
    x = Flatten()(x)
    x = Dense(1, activation='sigmoid')(x)

    model = Model([img, label], x)
    model.summary()

    return model

discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5), metrics=['accuracy'])
```

```python
def build_gan(generator, discriminator):
    noise = Input(shape=(100,))
    label = Input(shape=(len(y_train[0]),))
    img = generator([noise, label])

    discriminator.trainable = False
    valid = discriminator([img, label])

    combined = Model([noise, label], valid)
    combined.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5))

    return combined

gan = build_gan(generator, discriminator)
```

```python
def train(epochs, batch_size=128, save_interval=50):
    half_batch = int(batch_size / 2)

    for epoch in range(epochs):
        # Train Discriminator
        idx = np.random.randint(0, X_train.shape[0], half_batch)
        imgs = X_train[idx]
        labels = y_train[idx]

        noise = np.random.normal(0, 1, (half_batch, 100))
        gen_imgs = generator.predict([noise, labels])

        d_loss_real = discriminator.train_on_batch([imgs, labels], np.ones((half_batch, 1)))
        d_loss_fake = discriminator.train_on_batch([gen_imgs, labels], np.zeros((half_batch, 1)))
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train Generator
        noise = np.random.normal(0, 1, (batch_size, 100))
        sampled_labels = np.random.randint(0, y_train.shape[1], batch_size)
        sampled_labels = tf.keras.utils.to_categorical(sampled_labels, y_train.shape[1])

        valid_y = np.ones((batch_size, 1))

        g_loss = gan.train_on_batch([noise, sampled_labels], valid_y)

        # Print progress
        print(f"{epoch} [D loss: {d_loss[0]}, acc.: {100*d_loss[1]}%] [G loss: {g_loss}]")

def evaluate_discriminator():
    # Evaluate the discriminator on real and fake images
    num_samples = 100
    idx = np.random.randint(0, X_train.shape[0], num_samples)
    imgs = X_train[idx]
    labels = y_train[idx]

    noise = np.random.normal(0, 1, (num_samples, 100))
    gen_imgs = generator.predict([noise, labels])

    real_preds = discriminator.predict([imgs, labels])
    fake_preds = discriminator.predict([gen_imgs, labels])

    real_accuracy = np.mean(real_preds > 0.5)
    fake_accuracy = np.mean(fake_preds <= 0.5)

    print(f"Real Image Accuracy: {real_accuracy * 100:.2f}%")
    print(f"Fake Image Accuracy: {fake_accuracy * 100:.2f}%")

def save_imgs(epoch):
    r, c = 2, 5
    noise = np.random.normal(0, 1, (r * c, 100))
    sampled_labels = np.arange(0, len(y_train[0])).reshape(-1, 1)
    sampled_labels = tf.keras.utils.to_categorical(sampled_labels, len(y_train[0]))
    gen_imgs = generator.predict([noise, sampled_labels])

    gen_imgs = 0.5 * gen_imgs + 0.5

    fig, axs = plt.subplots(r, c)
    cnt = 0
    for i in range(r):
        for j in range(c):
            axs[i,j].imshow(gen_imgs[cnt])
            axs[i,j].set_title(f"Class: {np.argmax(sampled_labels[cnt])}")
            axs[i,j].axis('off')
            cnt += 1
    plt.show()

# Train the GAN
train(epochs=10000, batch_size=32, save_interval=200)

# Evaluate Discriminator
evaluate_discriminator()

# Save generated images after training
save_imgs(epoch=10000)  # Adjust the epoch parameter if needed
```

```python
# Define Models
def build_generator():
    input_noise = Input(shape=(100,))
    input_label = Input(shape=(1,))

    x = Dense(256 * 16 * 16, use_bias=False)(input_noise)
    x = Reshape((16, 16, 256))(x)

    x = Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    x = Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    x = Conv2D(3, (5, 5), padding='same', activation='tanh')(x)

    model = Model(inputs=[input_noise, input_label], outputs=x)
    return model

def build_discriminator():
    input_image = Input(shape=(64, 64, 3))

    x = Conv2D(64, (5, 5), strides=(2, 2), padding='same')(input_image)
    x = tf.keras.layers.LeakyReLU()(x)

    x = Conv2D(128, (5, 5), strides=(2, 2), padding='same')(x)
    x = tf.keras.layers.LeakyReLU()(x)

    x = Flatten()(x)
    x = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=input_image, outputs=x)
    return model

def build_cgan(generator, discriminator):
    discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5), metrics=['accuracy'])
    discriminator.trainable = False

    noise = Input(shape=(100,))
    label = Input(shape=(1,))
    generated_image = generator([noise, label])
    validity = discriminator(generated_image)

    model = Model(inputs=[noise, label], outputs=validity)
    model.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5))
    return model
```

```python
# Preprocess Dataset
def preprocess_image(image_path, img_height, img_width):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_png(image, channels=3)
    image = tf.image.resize(image, [img_height, img_width])
    image = (image / 127.5) - 1
    return image

def load_dataset(image_paths, img_height, img_width):
    images = [preprocess_image(path, img_height, img_width) for path in image_paths]
    return tf.data.Dataset.from_tensor_slices(images)
```

```python
def train_cgan(generator, discriminator, cgan, dataset, epochs, batch_size):
    for epoch in range(epochs):
        for batch in dataset.batch(batch_size):
            batch_size_actual = batch.shape[0]  # Get the actual batch size
            noise = np.random.randn(batch_size_actual, 100)
            labels = np.random.randint(0, 2, size=(batch_size_actual, 1))

            # Generate fake images
            generated_images = generator.predict([noise, labels])

            # Real images and labels for discriminator
            real_labels = np.ones((batch_size_actual, 1))
            fake_labels = np.zeros((batch_size_actual, 1))

            # Train discriminator on real and fake images
            d_loss_real = discriminator.train_on_batch(batch, real_labels)
            d_loss_fake = discriminator.train_on_batch(generated_images, fake_labels)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

            # Train generator
            g_loss = cgan.train_on_batch([noise, labels], real_labels)

        print(f'Epoch {epoch + 1}/{epochs} | D Loss: {d_loss[0]} | D Acc: {d_loss[1]} | G Loss: {g_loss}')
```

```python
# Generate and Save Images
def generate_and_save_images(generator, epoch, noise, labels, output_dir='/content/drive/MyDrive/SVG/generated_images'):
    generated_images = generator.predict([noise, labels])
    generated_images = (generated_images + 1) / 2.0

    for i in range(generated_images.shape[0]):
        plt.imshow(generated_images[i])
        plt.axis('off')
        plt.savefig(f'{output_dir}/generated_image_epoch_{epoch}_{i}.png')
        plt.close()
```

```python
# Evaluate the Model (Inception Score Example)
def calculate_inception_score(images, model, n_split=10):
    def get_predictions(images):
        processed_images = preprocess_input(images)
        predictions = model.predict(processed_images)
        return predictions

    splits = np.array_split(images, n_split)
    scores = []

    for split in splits:
        predictions = get_predictions(split)
        predictions = np.mean(predictions, axis=0)
        predictions = softmax(predictions, axis=-1)
        kl_divergence = np.sum(predictions * (np.log(predictions) - np.log(np.mean(predictions, axis=0))), axis=-1)
        scores.append(np.exp(np.mean(kl_divergence)))

    return np.mean(scores), np.std(scores)
```

```python
# Save and Load Models
def save_models(generator, discriminator, cgan):
    generator.save('generator_model.h5')
    discriminator.save('discriminator_model.h5')
    cgan.save('cgan_model.h5')

def load_models():
    generator = tf.keras.models.load_model('generator_model.h5')
    discriminator = tf.keras.models.load_model('discriminator_model.h5')
    cgan = tf.keras.models.load_model('cgan_model.h5')
    return generator, discriminator, cgan
```

```python
import os
from PIL import Image
import svgwrite

# Function to overlay logos on SVG hoodies
def overlay_logos(svg_folder, logo_folder, output_folder, logo_position):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Loop through all SVG files in the hoodie folder
    for svg_filename in os.listdir(svg_folder):
        if svg_filename.endswith('.svg'):
            hoodie_svg_path = os.path.join(svg_folder, svg_filename)

            # Create a new SVG drawing object for the output
            dwg = svgwrite.Drawing(filename=os.path.join(output_folder, f'modified_{svg_filename}'), size=("500px", "500px"))  # Adjust size as necessary

            # Read the SVG content and add it to the new SVG drawing
            with open(hoodie_svg_path, 'r') as file:
                hoodie_content = file.read()
                dwg.add(dwg.raw(hoodie_content))  # Using raw method correctly to add SVG content

            # Load the logo image
            logo_filename = 'your_logo.png'  # Change this to select the appropriate logo
            logo_path = os.path.join(logo_folder, logo_filename)

            # Open the logo image
            with Image.open(logo_path) as logo:
                # Resize the logo if needed
                logo = logo.resize((100, 100))  # Adjust size as necessary

                # Save the logo as a temporary PNG file
                logo_temp_path = os.path.join(output_folder, 'temp_logo.png')
                logo.save(logo_temp_path)

                # Add logo to the SVG drawing at the specified position
                dwg.add(dwg.image(logo_temp_path, insert=(logo_position[0], logo_position[1]), size=(100, 100)))

            # Save the modified SVG file
            dwg.save()

    print(f'Logos have been successfully added to SVG files in {output_folder}.')

# Example usage
svg_folder = '/content/drive/MyDrive/SVG/Hoodies'   # Replace with your folder path
logo_folder = '/content/drive/MyDrive/SVG/logo datasets/Logos'  # Replace with your folder path
output_folder = '/content/drive/MyDrive/SVG/ImposedImagesNew'    # Replace with your output folder
logo_position = (50, 50)  # Position where the logo will be placed (x, y)

overlay_logos(svg_folder, logo_folder, output_folder, logo_position)
```
